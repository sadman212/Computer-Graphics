```python
import matplotlib.pyplot as plt
# Function to plot 8-way symmetric points
def plot_circle_points(xc, yc, x, y, points):
    points.extend([
        (xc + x, yc + y), (xc - x, yc + y),
        (xc + x, yc - y), (xc - x, yc - y),
        (xc + y, yc + x), (xc - y, yc + x),
        (xc + y, yc - x), (xc - y, yc - x)
    ])
# Bresenhamâ€™s Circle Drawing Algorithm
def bresenham_circle(xc, yc, r):
    x, y = 0, r
    d = 3 - 2 * r
    points = []
    while x <= y:
        plot_circle_points(xc, yc, x, y, points)
        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        x += 1
    return points

def draw_circle(ax, xc, yc, r, color="blue"):
    points = bresenham_circle(xc, yc, r)
    xs, ys = zip(*points)
    ax.scatter(xs, ys, s=5, c=color)


```

#Task 1: Circles with Radii 5,8,12,20

```python
# Task 1
fig, ax = plt.subplots()
for r in [5, 8, 12, 20]:
    draw_circle(ax, 0, 0, r)
ax.set_title("Task 1: Circles with Radii 5,8,12,20")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```

#Task 2: Concentric Circles

```python
fig, ax = plt.subplots()
for r in [5, 10, 15]:
    draw_circle(ax, 0, 0, r)
ax.set_title("Task 2: Concentric Circles")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```

#Task 3: Shifted Circle

```python
fig, ax = plt.subplots()
draw_circle(ax, 30, 40, 15)
ax.set_title("Task 3: Shifted Circle at (30,40)")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```

#Task 4: Circle in Each Quadrant

```python
fig, ax = plt.subplots()
centers = [(20, 20), (-20, 20), (-20, -20), (20, -20)]
for (xc, yc) in centers:
    draw_circle(ax, xc, yc, 15)
ax.set_title("Task 4: Circles in Each Quadrant")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```

#Task 5: Filled Circle

```python
fig, ax = plt.subplots()
r = 15
for y in range(-r, r + 1):
    for x in range(-r, r + 1):
        if x*x + y*y <= r*r:
            ax.scatter(x, y, s=3, c="red")
ax.set_title("Task 5: Filled Circle")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```

#Task 6: Olympic Rings

```python
fig, ax = plt.subplots()
centers = [(0,0), (20,0), (40,0), (10,-15), (30,-15)]
colors = ["blue", "black", "red", "yellow", "green"]
for (c, col) in zip(centers, colors):
    draw_circle(ax, c[0], c[1], 12, color=col)

ax.set_title("Task 6: Olympic Rings")
ax.set_aspect('equal')
ax.axis("off")
plt.show()
```
