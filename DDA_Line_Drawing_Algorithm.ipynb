```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def int_round(n):
    return int(n + 0.5)
```

```python
def dda_line(x1, y1, x2, y2, color='blue'):
    dx = x2 - x1
    dy = y2 - y1

    x_points = []
    y_points = []

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for _ in range(steps + 1):
        x_points.append(int_round(x))
        y_points.append(int_round(y))
        x += x_inc
        y += y_inc

    plt.plot(x_points, y_points, marker='o', color=color)
```

```python
def draw_grid(max_x=50, max_y=50, step=1):
    for x in range(0, max_x + 1, step):
        plt.axvline(x, color='lightgray', linewidth=0.5)
    for y in range(0, max_y + 1, step):
        plt.axhline(y, color='lightgray', linewidth=0.5)
```

```python
def draw_triangle_with_grid():

    A = (5, 5)
    B = (25, 30)
    C = (40, 10)

    plt.figure(figsize=(8, 8))
    draw_grid(50, 50, step=5)

    dda_line(*A, *B, color='red')
    dda_line(*B, *C, color='green')
    dda_line(*C, *A, color='blue')

    for point in [A, B, C]:
        plt.plot(point[0], point[1], 'ko')

    plt.title("DDA Triangle with Integer Rounding and Grid")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.xlim(0, 50)
    plt.ylim(0, 50)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
draw_triangle_with_grid()
```
