```python
import matplotlib.pyplot as plt
import numpy as np
# ----- Helper Functions -----
def translate(points, tx, ty):
    """Translation of points by tx, ty"""
    T = np.array([[1, 0, tx],
                  [0, 1, ty],
                  [0, 0, 1]])
    return (T @ points.T).T
def scale(points, sx, sy):
    """Scaling of points by sx, sy"""
    S = np.array([[sx, 0, 0],
                  [0, sy, 0],
                  [0, 0, 1]])
    return (S @ points.T).T
def rotate(points, angle):
    """Rotation of points by given angle (in degrees)"""
    rad = np.radians(angle)
    R = np.array([[np.cos(rad), -np.sin(rad), 0],
                  [np.sin(rad), np.cos(rad), 0],
                  [0, 0, 1]])
    return (R @ points.T).T
# ----- Example Shape: A Triangle -----
triangle = np.array([[0, 0, 1],
                     [1, 0, 1],
                     [0.5, 1, 1],
                     [0, 0, 1]]) # closing the triangle

# ----- Apply Transformations -----
triangle_translated = translate(triangle, 2, 1)
triangle_scaled = scale(triangle, 2, 1.5)
triangle_rotated = rotate(triangle, 45)
# ----- Plot Results -----
plt.figure(figsize=(8, 8))
plt.plot(triangle[:,0], triangle[:,1], 'b-', label="Original")

plt.plot(triangle_translated[:,0], triangle_translated[:,1], 'r--', label="Translated")
plt.plot(triangle_scaled[:,0], triangle_scaled[:,1], 'g--', label="Scaled")
plt.plot(triangle_rotated[:,0], triangle_rotated[:,1], 'm--', label="Rotated")
plt.legend()
plt.grid(True)
plt.axis("equal")
plt.title("Geometric Transformations of a Triangle")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle (homogeneous coords)
triangle = np.array([
    [0, 0, 1],
    [1, 0, 1],
    [0.5, 1, 1],
    [0, 0, 1]
])

# Rotation matrix 45 degrees
angle = 45
rad = np.radians(angle)
R = np.array([
    [np.cos(rad), -np.sin(rad), 0],
    [np.sin(rad),  np.cos(rad), 0],
    [0, 0, 1]
])

# Matrix multiplication to get new vertices
rotated_triangle = (R @ triangle.T).T

print("New vertices after 45° rotation:")
print(rotated_triangle[:, :2])

# Plot original and rotated triangles
plt.figure(figsize=(6,6))
plt.plot(triangle[:,0], triangle[:,1], 'b-', label="Original Triangle")
plt.plot(rotated_triangle[:,0], rotated_triangle[:,1], 'r--', label="Rotated Triangle (45°)")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.title("Triangle Rotation by 45° about Origin")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define rectangle (homogeneous coords)
rectangle = np.array([
    [0, 0, 1],
    [2, 0, 1],
    [2, 1, 1],
    [0, 1, 1],
    [0, 0, 1]
])

# Scaling matrix (2x in x, 3x in y)
S = np.array([
    [2, 0, 0],
    [0, 3, 0],
    [0, 0, 1]
])

# Translation matrix (translate by (1,1))
T = np.array([
    [1, 0, 1],
    [0, 1, 1],
    [0, 0, 1]
])

# Scale then Translate
scaled_then_translated = (T @ (S @ rectangle.T)).T

# Translate then Scale
translated_then_scaled = (S @ (T @ rectangle.T)).T

# Print coordinates
print("Scale then Translate:\n", scaled_then_translated[:, :2])
print("\nTranslate then Scale:\n", translated_then_scaled[:, :2])

# Plotting
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(rectangle[:,0], rectangle[:,1], 'b-', label="Original")
plt.plot(scaled_then_translated[:,0], scaled_then_translated[:,1], 'r--', label="Scale then Translate")
plt.legend()
plt.title("Scale then Translate")
plt.axis('equal')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(rectangle[:,0], rectangle[:,1], 'b-', label="Original")
plt.plot(translated_then_scaled[:,0], translated_then_scaled[:,1], 'g--', label="Translate then Scale")
plt.legend()
plt.title("Translate then Scale")
plt.axis('equal')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle
triangle = np.array([
    [0, 0, 1],
    [1, 0, 1],
    [0.5, 1, 1],
    [0, 0, 1]
])

# Scaling matrix (2x)
S = np.array([
    [2, 0, 0],
    [0, 2, 0],
    [0, 0, 1]
])

# Rotation matrix (-30 degrees)
angle = -30
rad = np.radians(angle)
R = np.array([
    [np.cos(rad), -np.sin(rad), 0],
    [np.sin(rad),  np.cos(rad), 0],
    [0, 0, 1]
])

# Translation matrix (2, -1)
T = np.array([
    [1, 0, 2],
    [0, 1, -1],
    [0, 0, 1]
])

# Compose the transformations
M = T @ R @ S

# Apply combined transformation
transformed_triangle = (M @ triangle.T).T

# Output results
print("Composed Transformation Matrix (T * R * S):")
print(M)

print("\nTransformed Triangle Coordinates:")
print(transformed_triangle[:, :2])

# Plotting
plt.figure(figsize=(6,6))
plt.plot(triangle[:,0], triangle[:,1], 'b-', label='Original Triangle')
plt.plot(transformed_triangle[:,0], transformed_triangle[:,1], 'r--', label='Transformed Triangle')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.title("Combined Transformation: Scale, Rotate, Translate")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle (homogeneous coordinates)
triangle = np.array([
    [0, 0, 1],
    [1, 0, 1],
    [0.5, 1, 1],
    [0, 0, 1]
])

# Shear parameters
sh_x = 1.0  # shear in x-direction
sh_y = 0.5  # shear in y-direction

# Shear matrix combining both x and y shear
Shear = np.array([
    [1, sh_x, 0],
    [sh_y, 1, 0],
    [0, 0, 1]
])

# Apply shear transformation
sheared_triangle = (Shear @ triangle.T).T

# Print transformed coordinates
print("Original Triangle Coordinates:\n", triangle[:, :2])
print("\nSheared Triangle Coordinates:\n", sheared_triangle[:, :2])

# Plot original and sheared triangles
plt.figure(figsize=(6,6))
plt.plot(triangle[:,0], triangle[:,1], 'b-', label='Original Triangle')
plt.plot(sheared_triangle[:,0], sheared_triangle[:,1], 'r--', label='Sheared Triangle')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.title("Shear Transformation")
plt.show()
```
